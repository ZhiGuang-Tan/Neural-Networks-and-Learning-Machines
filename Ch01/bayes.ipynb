```python
import sys
sys.path.append("../util/")

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
#%matplotlib auto
```

```python
def bayes_classifier(dataset, mu1, mu2, Sigma):
    weights = np.linalg.solve(Sigma, mu1-mu2)
    bias = -0.5*np.dot(weights, mu1+mu2)
    return weights, bias
```
