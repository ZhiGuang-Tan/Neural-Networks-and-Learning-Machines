```python
import numpy as np
import matplotlib.pyplot as plt
#get_ipython().run_line_magic('matplotlib', 'auto')
import matplotlib.cm as cm
```

```python
def double_moon(n, w=6, d=1, r=10):
    dataset = np.zeros([n, 3])
    rho = np.random.uniform(r-0.5*w, r+0.5*w, n)
    theta = np.random.uniform(0, 2*np.pi, n)
    for i in range(n):
        dataset[i, 0] = rho[i]*np.cos(theta[i])
        dataset[i, 1] = rho[i]*np.sin(theta[i])
        if ((theta[i]>0.0*np.pi) & (theta[i]<1.0*np.pi)):
            dataset[i, 2] = +1
        else:
            dataset[i, 0] += r
            dataset[i, 1] -= d
            dataset[i, 2] = -1
    return dataset
```

```python
def gaussian_2d(mu1, mu2, Sigma):
    tmp = np.random.multivariate_normal(mu1, Sigma, 100)
    tmp = np.concatenate((tmp, np.zeros((100, 1))), axis=1)
    tmp1 = np.random.multivariate_normal(mu2, Sigma, 100)
    tmp1 = np.concatenate((tmp1, np.ones((100, 1))), axis=1)
    tmp = np.concatenate((tmp, tmp1), axis=0)
    np.random.shuffle(tmp)
    return tmp
```
